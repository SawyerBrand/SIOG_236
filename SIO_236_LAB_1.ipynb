```python
# We want to be able to see any plot we make in the actual notebook so we write:
%matplotlib inline
# We want the numpy library in order to use sin,cos,etc 
import numpy as np
from numpy import pi, sin, cos, arccos, arcsin, arctan2, sqrt
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
# Inclination of whichever satellite you want to use (I used TOPEX)
i = 66.01*pi/180  
#satellite period of whichever satellite you want to use 
T = 6745.72 
we = 2*pi/86164.1 # Earth's orbital frequency
ws = 2*pi/T # Satelite's orbital frequency 
wn = we - ws*10/127 # precession frequency
we_prime = we - wn
frac = 21
t = np.arange(0,frac*T,1)

theta = arcsin(sin(ws*t)*sin(i))
top = -sin(we_prime*t)*cos(ws*t) + cos(we_prime*t)*sin(ws*t)*cos(i)
bot = cos(we_prime*t)*cos(ws*t)+sin(we_prime*t)*sin(ws*t)*cos(i) 
phi = arctan2(top,bot) # longitude in radians

# Converting back to degrees
lon = phi*180/pi
lat = theta*180/pi

plt.figure(figsize=(20,20))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()
plt.plot(lon,lat,'.',c='red',markersize=.5)
plt.show()
```
